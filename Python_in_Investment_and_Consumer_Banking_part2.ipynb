#### <FONT COLOR ="ROYALBLUE">MOVING AVERAGES IN TRADING

A moving average is a widely-used technical analysis tool that helps traders to smooth out price data by creating a constantly updated average price. It essentially filters out the noise in the market, making it easier to identify trends and potential buying or selling opportunities.

Moving Average 10 (MA10): This is the average of the closing prices of a financial instrument over the past 10 days. It's a short-term indicator that reacts quickly to price changes.
Moving Average 50 (MA50): This is the average of the closing prices over the past 50 days. It's a longer-term indicator that responds more slowly to price changes.
Random Variables

In statistics, a random variable is a variable whose possible values are determined by the outcome of a random event. In the context of moving averages, we can think of the daily closing prices as random variables, as their values are influenced by unpredictable factors like market sentiment, economic news, or corporate announcements.

#### <FONT COLOR ="ROYALBLUE"> PROBABILITY RULES

The relationship between the MA10 and MA50 and make informed trading decisions, it's helpful to be familiar with basic probability rules:

The Probability Rule: The probability of an event occurring is between 0 and 1, inclusive. A probability of 0 means the event will never occur, while a probability of 1 means it's certain to happen.
The Complementary Rule: The probability of an event not occurring is equal to 1 minus the probability of the event occurring (P(A') = 1 - P(A)).
The Addition Rule: If two events are mutually exclusive (cannot both happen at the same time), the probability of either event occurring is the sum of their individual probabilities (P(A or B) = P(A) + P(B)).
The Multiplication Rule: If two events are independent (the outcome of one event does not affect the outcome of the other), the probability of both events occurring is the product of their individual probabilities (P(A and B) = P(A) * P(B)).
Applying Probability Rules to the Trading Strategy

By understanding these probability rules, traders can analyze the likelihood of different scenarios happening based on historical data. For instance, traders might want to determine the probability of the MA10 crossing above the MA50, which could signal a potential buying opportunity, or the probability of the MA10 crossing below the MA50, indicating a potential selling opportunity.

By incorporating probability rules into their analysis, traders can make more informed decisions about when to enter or exit a position based on the moving averages. However, it's essential to remember that past performance does not guarantee future results, and any trading strategy should be used in conjunction with other technical and fundamental analysis tools.

Random variables and outcomes are fundamental concepts in probability and statistics. Let's break down each term and provide an example to illustrate their meanings.

1. Random Variable:

A random variable is a variable that takes on different values based on the outcomes of a random event or experiment. It assigns a numerical value to each outcome in the sample space, which is the set of all possible outcomes. Random variables can be either discrete or continuous, depending on the nature of the possible values they can take.

2. Example:

Let's consider a simple random experiment: flipping a fair coin. We can define a discrete random variable X to represent the number of heads obtained after flipping the coin once. The possible values of X are 0 and 1, corresponding to the two possible outcomes: tails (0 heads) and heads (1 head).

3. Outcome:

An outcome is a single, specific result of a random event or experiment. It is one of the many possible results that can occur when the random event takes place. In the context of a random variable, each outcome is assigned a numerical value.

4. Example:

Continuing with the coin-flipping experiment, there are two possible outcomes when flipping a fair coin:

Heads
Tails
When we defined the random variable X to represent the number of heads obtained, we assigned a numerical value to each outcome:

If the outcome is Heads, the value of X is 1.
If the outcome is Tails, the value of X is 0.
In this example, Heads and Tails are the two possible outcomes of the random event (flipping a coin), and the random variable X maps these outcomes to their corresponding numerical values (1 and 0, respectively).

### <font color="red">3. FREQUENCY

Frequency in the context of random variables refers to the number of times a particular value or outcome occurs in a dataset or a series of trials of a random experiment. It is a measure of how often a specific value appears and can help identify patterns or trends in the data.

In probability and statistics, the concept of frequency is often used to estimate probabilities or to create probability distributions, particularly for discrete random variables.

Let's illustrate this concept with an example:

Suppose we roll a fair six-sided die 100 times and record the outcomes. The dataset might look like this:

[1, 6, 3, 4, 2, 1, 6, 4, 3, 5, n]

To calculate the frequency of each outcome, we count the number of times each value (1-6) appears in the dataset:

The number 1 appears 15 times.

The number 2 appears 20 times.

The number 3 appears 18 times.

The number 4 appears 17 times.

The number 5 appears 12 times.

The number 6 appears 18 times.

These frequencies give us an empirical distribution of the die rolls. To estimate the probabilities associated with each outcome, we divide the frequency of each value by the total number of trials (in this case, 100):

P(1) ≈ 15/100

P(2) ≈ 20/100

P(3) ≈ 18/100

P(4) ≈ 17/100

P(5) ≈ 12/100

P(6) ≈ 18/100

By analyzing the frequency of each value in the dataset, we can gain insights into the underlying probability distribution of the random variable and make inferences about the population or the random process being studied.

### <font color="red">3. RELATIVE FREQUENCY

Relative frequency is a measure used in statistics to express the proportion of times a particular outcome or value occurs in a dataset or a series of trials of a random experiment. It is calculated by dividing the frequency (the number of times the outcome occurs) by the total number of trials or observations in the dataset.


Relative frequency is a way to estimate probabilities based on observed data, particularly for discrete random variables. It provides a normalized representation of the data, where the sum of the relative frequencies of all possible outcomes equals 1.


Let's revisit the example of rolling a fair six-sided die 100 times, with the following frequencies for each outcome:

The number 1 appears 15 times.

The number 2 appears 20 times.

The number 3 appears 18 times.

The number 4 appears 17 times.

The number 5 appears 12 times.

The number 6 appears 18 times.

To calculate the relative frequency for each outcome, we divide the frequency of each value by the total number of trials (100):

Relative frequency of 1 = 15/100 = 0.15

Relative frequency of 2 = 20/100 = 0.20

Relative frequency of 3 = 18/100 = 0.18

Relative frequency of 4 = 17/100 = 0.17

Relative frequency of 5 = 12/100 = 0.12

Relative frequency of 6 = 18/100 = 0.18

These relative frequencies represent an empirical probability distribution of the die rolls. They indicate the proportion of times each outcome occurred in the 100 trials and can be used to estimate the underlying probabilities of the random variable. The sum of all relative frequencies is equal to 1, as expected for a probability distribution.

### <font color="red">6. CONTINOUS RANDOM VARIABLES

A distribution of a continuous variable refers to the way in which the values of a particular continuous variable are spread across its range. Continuous variables are variables that can take on any value within a defined range, as opposed to discrete variables, which can only take on a specific set of values.

The distribution of a continuous variable is often described using a probability density function, which defines the probability of observing a particular value within a certain interval. This function is a smooth curve that represents the overall shape of the distribution. There are several commonly used distributions to describe continuous variables, including the normal distribution, the exponential distribution, and the uniform distribution.

For example, the normal distribution is often used to describe the distribution of a continuous variable that is symmetrical and bell-shaped, with a single peak and tails that taper off smoothly. On the other hand, the exponential distribution is often used to describe the distribution of a continuous variable that is positively skewed and has a single peak at zero.

Probability Density Function & Cumulative Distribution Function

Probability Density Function (pdf) and Cumulative Distribution Function (cdf) are two important concepts in probability and statistics.

A Probability Density Function is a function that describes the likelihood of a random variable taking on a given value. For continuous variables, the pdf gives the density of the variable at each point in its range. The pdf can be thought of as a smooth curve that describes the distribution of the variable. The total area under the pdf curve is equal to 1, which represents the total probability of all possible values of the random variable.

A normal distribution with mean 0 and standard deviation 1.

 f(x) = (1 / sqrt(2 * pi)) * e^(-x^2 / 2)


The pdf curve for this normal distribution is symmetrical and bell-shaped. The height of the curve at any given point represents the likelihood of the random variable taking on that value.

A Cumulative Distribution Function, on the other hand, is a function that describes the cumulative probability of a random variable being less than or equal to a given value. The cdf is a monotonically increasing function that ranges from 0 to 1. The cdf value at any given point represents the total probability of the random variable being less than or equal to that value.

consider the same normal distribution with mean 0 and standard deviation 1. The cdf of this normal distribution can be calculated using the cumulative distribution function of the standard normal distribution

F(x) = (1 / 2) * (1 + erf(x / sqrt(2)))


```python
from scipy.stats import norm
```

```python
density = pd.DataFrame()
density['x'] = np.arange(-4, 4, 0.001)
density['pdf'] = norm.pdf(density['x'], 0, 1)
density['cdf'] = norm.cdf(density['x'], 0, 1)
```

Creating a pandas DataFrame called density and calculates the probability density function (pdf) and cumulative distribution function (cdf) values for the normal distribution with mean 0 and standard deviation 1.

The density DataFrame has two columns: x and pdf, and cdf. The x column contains an array of values ranging from -4 to 4, with a step size of 0.001. The pdf column contains the values of the probability density function for each x value, and the cdf column contains the values of the cumulative distribution function for each x value.

The norm.pdf function calculates the pdf values, and the norm.cdf function calculates the cdf values. Both functions take three arguments: the data points for which the pdf or cdf values should be calculated, the mean of the normal distribution, and the standard deviation of the normal distribution. In this case, the data points are the x values from the density DataFrame, the mean is 0, and the standard deviation is 1.

  **Value at Risk (VaR)**

Value at Risk (VaR) is a statistical measure used to quantify the risk associated with an investment. It is defined as the maximum loss that an investment is expected to experience at a given confidence level within a specific time horizon. VaR is used to determine the amount of capital that should be set aside to cover potential losses.

```python
# Define the mean and standard deviation of the daily log return
mu = aapl['LogReturn'].mean()
sigma = aapl['LogReturn'].std()

# Calculate the VaR at a confidence level of 95%
var = norm.ppf(0.05, mu, sigma)

print("The Value at Risk (VaR) at a 95% confidence level is: ", var)
```
